```python
def solution(id_list, report, k):
    answer = [0] * len(id_list)
    cntReportedByUserId = {idx: 0 for idx in id_list}
    report = set(report)
    
    for item in report:
        cntReportedByUserId[item.split()[1]] += 1;

    for item in report:
        reporter, reported = item.split()
        if cntReportedByUserId[reported] >= k:
            answer[id_list.index(reporter)] += 1
        
    return answer
```

```python
id_list = ["muzi", "frodo", "apeach", "neo"]
report = ["muzi frodo","apeach frodo","frodo neo","muzi neo","apeach muzi"]
k = 2
```

```python
solution(id_list, report, k)
```
